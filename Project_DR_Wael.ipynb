```python
df['salary'] = df['salary'].str.strip()
df['salary'].replace({'<=50K':0,'>50K':1}, inplace=True)
```

```python
df.select_dtypes(include=['int','float']).corr()['salary']
```

```python
df.columns=df.columns.str.lower()
```

```python
cols_to_fill = ['workclass', 'occupation', 'native-country']

for col in cols_to_fill:
    df[col] = df[col].str.strip()
    mode_value = df[col][df[col] != '?'].mode()[0]

    df[col] = df[col].replace('?', mode_value)
```

```python
df['occupation'].value_counts()
```

```python
df.tail()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['overtime'] = df['hours-per-week'].apply(lambda x: 1 if x > 40 else 0)
df['net_capital'] = df['capital-gain'] - df['capital-loss']
```

```python
df['sex'] = df['sex'].astype(str).str.strip()
df['sex'] = df['sex'].replace('?', None)
df['sex'] = df['sex'].map({'Male': 0, 'Female': 1})

categorical_cols = ['native-country', 'occupation', 'marital-status', 'workclass', 'relationship', 'race']

df_encoded = df.copy()

for col in categorical_cols:
    unique_values = df[col].dropna().unique()
    for val in unique_values:
        df_encoded[f"{col}_{val}"] = (df[col] == val).astype(int)


```

```python
from sklearn.model_selection import train_test_split


X = df_encoded.drop(columns=['salary', 'workclass', 'occupation', 'marital-status',
                             'relationship', 'race', 'native-country', 'education'])

Y=df_encoded['salary']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=50, random_state=42,min_samples_leaf=2,max_depth=14)
model.fit(X_train, Y_train)
```

```python
ypred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(Y_test,ypred)
```
